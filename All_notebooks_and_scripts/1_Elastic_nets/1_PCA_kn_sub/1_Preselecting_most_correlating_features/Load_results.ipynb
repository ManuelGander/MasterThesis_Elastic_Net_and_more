```python
frames = []
for k in tqdm(keys):
    df = pd.read_pickle(mypath+'/'+k)
    frames.append(df)
Df = pd.concat(frames, ignore_index=True)
print(Df['dataset'].iloc[0])
print(Df['source'].iloc[0])
del Df['dataset']
del Df['source']
```

```python
df = Df.groupby(['PubChem_CID', 'alpha', 'l1_ratio', 'kn']).mean().reset_index()
df = df[np.isfinite(df['pearsons'])]
df['drug_repr'] = df['PubChem_CID'].map(D_dr_rep)
df['RMSE_ratio'] = df['RMSE']/df['RMSE_mean_model']
df.sort_values('pearsons')
```

```python
# Pick the one with the best l1_ratio and alpha
crit = 'RMSE_ratio'
crit = 'pearsons'

alphas = sorted(set(df['alpha']))
l1_ratios = sorted(set(df['l1_ratio']))
kns = sorted(set(df['kn']))

frames = []
drugs = sorted(set(df['PubChem_CID']))
for kn in kns:
    for d in drugs:
        dfss = df[(df['PubChem_CID']==d)&(df['kn']==kn)]
        if crit=='RMSE_ratio':
            frames.append(dfss.sort_values(crit).iloc[:1])
        else:
            frames.append(dfss.sort_values(crit).iloc[-1:])
dfm = pd.concat(frames, ignore_index=True)
dfm.sort_values('pearsons')
```

```python
x_data = list(Dft['kn'])
y_data = list(Dft['td'])
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Define the second-order polynomial function
def polynomial(x, a, b, c):
    return a * x**2 + b * x + c

# Use curve_fit to find the best fit parameters a, b, and c
params, covariance = curve_fit(polynomial, x_data, y_data)

# Extract the parameters
a, b, c = params
print(f"Fitted parameters: a={a}, b={b}, c={c}")

# Generate y values using the fitted polynomial
xs = np.linspace(0,20000,10001)
y_fit = polynomial(xs, a, b, c)

# Plot the original data and the fitted curve
plt.scatter(x_data, y_data, label='Data', s=1)
plt.plot(xs, y_fit, color='red', label='Fitted curve')
plt.legend()
plt.xlabel('# of highest correlating features selected')
plt.ylabel('Running Time per Iteration [s]')
plt.title('Second-Order Polynomial Fit - NCI60 Drugscreen Data')
plt.show()
```
