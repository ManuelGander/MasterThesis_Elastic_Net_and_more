```python
D_cid_classes={}
for iii,cid in enumerate(cids):
    print(iii)
    keys=list(D_cid_classes.keys())
    if cid in keys:
        parent_cid=D_cid_classes[cid]
    else:
        parent_cid=cid
    
    # Get all possible drug names:
    possible_names=list(dfn[dfn['PubChem_CID']==cid]['Drug'])
    try: 
        cid=int(cid)
        possible_names=possible_names+list(df[df[0]==cid][1])
    except:
        pass
        
    # Get all corresponding cids
    alternative_cids0=list(set(df[df[1].isin(possible_names)][0]))
    alternative_cids1=[]
    alt2=list(dfn[dfn['Drug'].isin(possible_names)]['PubChem_CID'])
    for cid in alt2:
        try: 
            cid=int(cid)
            alternative_cids1.append(cid)
        except:
            alternative_cids1.append(cid)
    
    alternative_cids=list(set(alternative_cids0+alternative_cids1))
    
    for cid in alternative_cids:
        D_cid_classes[cid]=parent_cid
    # Mmap them to this parent cid
```
