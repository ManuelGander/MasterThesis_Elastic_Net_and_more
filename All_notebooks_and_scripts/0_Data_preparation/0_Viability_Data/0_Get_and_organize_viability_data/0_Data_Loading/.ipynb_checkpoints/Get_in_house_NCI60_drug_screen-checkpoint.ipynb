```python
signal_quality_bitstring=""                                                                                                                          
dose1_bitstring=""                                                                                                         
dose2_bitstring=""
dose3_bitstring=""
dose4_bitstring=""
dose5_bitstring=""
```

```python
import base64
def bit_string_to_array(encoded_data):
    decoded_data = base64.b64decode(encoded_data)
    array = np.frombuffer(decoded_data, dtype=np.float64)
    shape = (10114,)
    array = array.reshape(shape)
    return(array)
```

```python
sq=bit_string_to_array(signal_quality_bitstring)
d1=bit_string_to_array(dose1_bitstring)
d2=bit_string_to_array(dose2_bitstring)
d3=bit_string_to_array(dose3_bitstring)
d4=bit_string_to_array(dose4_bitstring)
d5=bit_string_to_array(dose5_bitstring)
```

```python
ccl_drug_combination=pd.read_csv(f'{Path}/NCI60_screen.csv', header=None)[0]
ccl_drug_combination=[a.replace(' ', '')[1:-2] for a in ccl_drug_combination]
ccls=[a.split('_')[0] for a in ccl_drug_combination]
drugs=[a.split('_', 1)[1] for a in ccl_drug_combination]
```

```python
df=pd.DataFrame({'CCL':ccls, 'Drug':drugs, 'Dose0':np.ones_like(d1), 'Dose1':d1, 'Dose2':d2, 'Dose3':d3,
                'Dose4':d4, 'Dose5':d5, 'Signal_Quality':sq})
```

```python
frames=[]

for i in range(5):
    dfs=df[['CCL', 'Drug', f'Dose{i+1}']].copy()
    dfs.columns=['CCL', 'Drug', 'Viability']
    dfs['Dose']=10**(i-3)
    frames.append(dfs)
```

```python
df=pd.concat(frames, ignore_index=True)
```

```python
M=df.copy()
```

```python
M['CCL']=M['CCL'].astype('category')
M['Drug']=M['Drug'].astype('category')
C=pd.crosstab(M['CCL'], M['Drug'])
```

```python
for i in range(100):
    cs=C.sample().T.sample(1)
    if cs.values>3:
        dr=list(cs.index)[0]
        ccl=list(cs.columns)[0]
        Ms=M[(M['CCL']==ccl)&(M['Drug']==dr)].copy()
        sns.lineplot(data=Ms, x='Dose', y='Viability')
        break
plt.ylim(0,2)
plt.xscale('log')
```
