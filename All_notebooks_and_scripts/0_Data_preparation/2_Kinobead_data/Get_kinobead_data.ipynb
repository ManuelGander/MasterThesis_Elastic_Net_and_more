```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
from os import listdir
from os.path import isfile, join, isdir
```

```python
def extract_intensities(df):
    id_cols=['Protein IDs', 'Majority protein IDs', 'Protein names', 'Gene names']
    concs=[a.split( )[-1] for a in df.columns if 'Razor + unique peptides' in a and 'nM' in a]
    dfc=pd.DataFrame({'conc':concs})
    dfc['nM']=[float(a.split('nM')[0]) for a in concs]
    dfc=dfc.sort_values('nM')
    
    dfc=pd.concat([pd.DataFrame({'conc':['DMSO'], 'nM':[0.1]}), dfc], ignore_index=True)
    dfc['LFQ']='LFQ intensity '+ dfc['conc']
    dfc['Raw']='Intensity '+ dfc['conc']

    dfs=df[[*id_cols, *list(dfc['LFQ']), *list(dfc['Raw'])]].copy()
    return(dfs, dfc)
```

```python
D_frames={}

mypath='/media/kusterlab/internal_projects/active/Drugscreens/ClinicalKIs2.0/DrugScreenMSData/'
dirs = [f for f in listdir(mypath) if isdir(join(mypath, f))]

for i in range(len(dirs)):
    directory=dirs[i]
    dirpath=mypath+directory

    ms_dir=directory+'/combined'

    if 'combined' in listdir(dirpath):
        df=pd.read_csv(mypath+'/'+directory+'/'+'combined/txt/proteinGroups.txt', sep='\t')
        dfs, dfc=extract_intensities(df)
        dfs['Drug']=directory
        key = ', '.join(list(dfc['conc']))
        if not key in list(D_frames.keys()):
            D_frames[key]=[]
        D_frames[key].append(dfs)
    else:
        print(f'Missing: {directory}')
```
