```python
def select_intensity(Df, key):
    # I keep LFQ if I have it for DMSO and for the first dose.
    # This reproduces the .pdf's almost always, i.e. for some rare cases I do LFQ where the pdf does raw, but I haven't observed that 
    # for the nicely regulated ones, jsut for some random ones
    
    conc1=key.split(', ')[1]
    lfq_cols=[a for a in Df.columns if 'LFQ' in a]
    int_cols=[a for a in Df.columns if 'Intensity' in a]
    
    lfq_inds=np.where((Df[['LFQ intensity DMSO', 'LFQ intensity '+conc1]]>0).sum(1)==2)[0]
    int_inds=np.where((Df[['LFQ intensity DMSO', 'LFQ intensity '+conc1]]>0).sum(1)!=2)[0]
    
    Df_lfq=Df.iloc[lfq_inds].copy()
    Df_lfq['LFW_or_raw']='lfq'
    Df_int=Df.iloc[int_inds].copy()
    Df_int[lfq_cols]=Df_int[int_cols]
    Df_int['LFW_or_raw']='raw'
    Dfs=pd.concat([Df_lfq, Df_int], ignore_index=True)
    for c in int_cols:
        del Dfs[c]
    return(Dfs)
```

```python
Dfs=select_intensity(Df, key)
Dfs['key']=Dfs['Drug']+'_'+Dfs['Protein IDs']+'_'+Dfs['Gene names']
```
