```python
import pandas as pd
import numpy as np
import seaborn as sns
import scipy
import matplotlib.pyplot as plt
import torch
import time
```

```python
sns.set(rc={'figure.figsize':(5,5)})
sns.set_style('whitegrid')
plt.rcParams['figure.dpi'] = 300
```

```python
Path='/home/mgander/Atlantic/data/'
```

```python
M=pd.read_pickle(f'{Path}copied_data/Atlantic/New_MQ_ph_M_dup.pkl')
M=np.log10(M).copy()
M
```

```python
dups=['HCT116', 'SW620', 'HT29']
```

# Using the filled (nans only) data

```python
#phos_peps=[a for a in M.index if a[0]=='P']
#M=M.loc[phos_peps].copy()

# Remove rare peptides 
fr=0.9
M=M[np.isnan(M).sum(1)<(1-fr)*len(M.columns)].copy()
# z-transform each pepitde
```

```python
M=(M.T-np.nanmean(M, 1)).T
M=(M.T/np.nanstd(M, 1)).T
```

```python
seqs=[a for a in M.index if '(Phospho' in a]
```

```python
M=M.loc[seqs].copy()
```

```python
M.index=[a.split('(Phospho (STY))')[0]+'p'+a.split('(Phospho (STY))')[1] for a in M.index]
```

```python
palette = sns.color_palette('tab10')
```

```python
from sklearn.decomposition import PCA


# Perform PCA
pca = PCA(n_components=2)  # Set the number of components to 2 for 2D PCA
transformed_data = pca.fit_transform(np.nan_to_num(M.values, 0).T)

dfs=pd.DataFrame({'x':transformed_data[:,0], 'y':transformed_data[:,1], 'cellline':M.columns})
dfs['cellline2']=[a if a!='Hela' else 'Hela_Hela' for a in dfs['cellline']]
dfs['Set']=[a.split('_')[1] for a in dfs['cellline2']]
sns.scatterplot(data=dfs, x='x', y='y', hue='Set', s=70, palette=palette)
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
```

```python
dfs['Set']=[a.split('_')[0] if a.split('_')[0] in dups else a.split('_')[1] for a in dfs['cellline2'] ]
Ds={}
Ds['NCI60']=0
Ds['CRC65']=1
Ds['HCT116']=2
Ds['HT29']=3
Ds['SW620']=4
Ds['Hela']=5
dfs['zs']=[Ds[a] for a in dfs['Set']]
dfs=dfs.sort_values('zs')
dfs.index=range(len(dfs))
```

```python
sns.scatterplot(data=dfs, x='x', y='y', hue='Set', s=70, palette='tab10')
plt.legend(bbox_to_anchor=(0.75, 1), loc='upper left')
plt.xlabel("PCA 1")
plt.ylabel("PCA 2")
```

```python
pca = PCA(n_components=5)  # Set the number of components to 2 for 2D PCA
transformed_data = pca.fit_transform(np.nan_to_num(M.values, 0).T)
Df=pd.DataFrame(data=transformed_data.T, columns=M.columns)
Df.to_pickle(f'{Path}Representations/basic_PCA_5.pkl')
Df
```
