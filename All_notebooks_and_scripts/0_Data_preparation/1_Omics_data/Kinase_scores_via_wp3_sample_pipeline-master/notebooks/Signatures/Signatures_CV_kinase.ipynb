```python
# needed functions and libraries
import pandas as pd
import seaborn as sns
import signature_functions as signature_functions
```

# KINASE Signatures

```python
# preprocess
import warnings
entity_subtypes = 'Histologic subtype' # meta_data column it should be also the same in the signature functions
warnings.filterwarnings('ignore')
report_directory = '/media/kusterlab/internal_projects/active/TOPAS/WP31/Playground/Retrospective_study/2023.04.27_CJ_final_cohort_reviewed'
meta_file_path = '/media/kusterlab/internal_projects/active/TOPAS/WP31/Playground/Retrospective_MTBs_Evaluation/SOLID_TUMORS_Metadata_230427_CJ.xlsx'
kinase_results_folder = 'kinase_results'



#ROC_curve_threshold = 0.8
p_value_threshold = 0.05
meta_data = pd.read_excel(meta_file_path)
meta_data = meta_data[~meta_data[entity_subtypes].str.contains("\?",na=True)]

kinase_scores_file = f'{report_directory}/{kinase_results_folder}/kinase_scores.tsv'
kinase_scores_file_annotation = f'{report_directory}/{kinase_results_folder}/scored_peptides.tsv'
```

```python
# select from the list bellow
list_entities=meta_data[entity_subtypes].unique().tolist()
for interseted_entity in list_entities:
    print(f'Runnging the siganture for the {interseted_entity}')
    kinase_scores = pd.read_csv(kinase_scores_file,sep='\t')
    kinase_scores_annotation_df = pd.read_csv(kinase_scores_file_annotation,sep='\t')
    annotation_file = pd.read_excel(meta_file_path)
    interested_patinets = annotation_file['Sample name'][annotation_file[entity_subtypes] == interseted_entity].tolist()
    kinase_scores.index = kinase_scores['PSP Kinases']
    kinase_scores_df = kinase_scores.transpose()
    all_patinets =  annotation_file['Sample name'].unique().tolist()
    kinase_scores_df = kinase_scores_df[kinase_scores_df.index.isin(all_patinets)]
    kinase_list = kinase_scores_df.columns
    kinase_scores_df['Entity'] = 'others'
    kinase_scores_df['Entity'][kinase_scores_df.index.isin(interested_patinets)] = [interseted_entity]

    # calculation of the singnatures 
    t_result = signature_functions.one_vs_all_t_test(kinase_scores_df,kinase_list, interseted_entity,'Entity')
    new_sginatures = signature_functions.get_final_signatures_from_t_test_results_by_criteria(t_result,p_value_threshold=p_value_threshold,
                                                                                            fdr_threshold=None,delta_threshold=1)
    if len(new_sginatures.index > 0):
        print(f'now saving the singature files for {interseted_entity} in  {kinase_results_folder} ')
        new_sginatures.to_excel(f'{report_directory}/{kinase_results_folder}/{interseted_entity}_vs_others_pvalue_{p_value_threshold}.xlsx')
```
