# Simple Kinase Scoring

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
```

```python
results_folder = '/media/kusterlab/internal_projects/active/TOPAS/WP31/Playground/Retrospective_study/2024.01.11_CJ_pancancer_166'
# results_folder = '/put/it/here/'
patient_regex = "^\\S+-.+-\\S+"

patients_proteins = pd.read_csv(Path(results_folder) / 'annot_pp.csv', index_col=0, nrows=1)
patcols = patients_proteins.filter(regex=patient_regex).columns.tolist()
patcols = {colname: 'float32' for colname in patcols}
usecols = ['Modified sequence', 'Gene names', 'Proteins', 'Site positions identified (MQ)', 'PSP Kinases'] + list(patcols.keys())
patients_proteins = pd.read_csv(Path(results_folder) / 'annot_pp.csv', index_col=0, dtype=patcols, usecols=usecols)
```

```python
patients_proteins
```

```python
# Median normalization
min50 = patients_proteins.filter(regex=patient_regex).count(axis=1) > 0.5 * len(patients_proteins.filter(regex=patient_regex).columns)
min50 = patients_proteins.loc[min50]
medians = min50.filter(regex=patient_regex).median(axis=0)
medians
```

```python
medmean = np.mean(medians)
factors = medians / medmean
```

```python
patients_proteins.loc[:, list(patcols.keys())] = patients_proteins.filter(regex=patient_regex).apply(lambda x: x * factors[x.name], axis=0)
```

### Phosphosite scoring: Mean of log intensities of all kinase substrates

Split-explode approach for each PSP kinase entry; sites will be count multiple times if they are phosphorylated by multiple kinases

```python
prot = patients_proteins.copy()

# split-explode the PSP Kinase column
prot['PSP Kinases'] = prot['PSP Kinases'].str.split(';')
prot = prot.explode('PSP Kinases')
```

```python
prot.groupby('PSP Kinases').agg({patient: np.mean for patient in prot.filter(regex=patient_regex).columns.tolist()})
```

Unique peptide approach; sites targeted by multiple kinases will not be considered at all

```python
prot = patients_proteins.copy()

# filter out peptides potentially phosphorylated by multiple kinases
prot = prot.loc[prot['PSP Kinases'].str.count(';') == 0]
```

```python
prot.groupby('PSP Kinases').agg({patient: np.mean for patient in prot.filter(regex=patient_regex).columns.tolist()})
```

Occam's razor approach; sites targeted by multiple kinases will be considered only once, for the kinase with the highest number of unique peptides. Kinases showing no unique peptides will still be ignored.

```python
prot = patients_proteins.copy()

# count the number of unique peptides per kinase
kinases = prot['PSP Kinases']
kinases = kinases.loc[kinases.str.count(';') == 0]
kinases = kinases.value_counts()
```

```python
def find_most_occurring_kinase(kinase_string):
    if type(kinase_string) != str:
        raise ValueError(f'Unexpected value {kinase_string} of type {type(kinase_string)}')
    kinase_string = kinase_string.split(';')
    kinase_string = [kinase for kinase in kinase_string if kinase in kinases.index]
    if len(kinase_string) == 0:
        return np.NaN
    else:
        max = 0
        for kinase in kinase_string:
            if kinases[kinase] > max:
                max = kinases[kinase]
                max_kinase = kinase
        return max_kinase

amb = prot['PSP Kinases']
amb = amb.loc[amb.str.count(';') != 0]
amb = amb.loc[~amb.isna()]
pre_dict = pd.concat([amb, amb.apply(find_most_occurring_kinase).rename('into')], axis=1).drop_duplicates()
repl_dict = pd.Series(pre_dict['into'].values, index=pre_dict['PSP Kinases']).to_dict()
```

```python
prot['PSP Kinases'] = prot['PSP Kinases'].replace(repl_dict)
```

```python
prot.groupby('PSP Kinases').agg({patient: np.mean for patient in prot.filter(regex=patient_regex).columns.tolist()})
```

### Kinase abundance scoring: Mean of log intensities of all peptides belonging to a kinase

```python
all_kinases = pd.read_csv('/media/kusterlab/line_functions/bioinformatics/Databases/psite_annotation_mapping/PhosphoSitePlus/Kinase_Substrate_Dataset', sep='\t', skiprows=2)
all_kinases = all_kinases.loc[all_kinases['GENE'].notna()]
all_kinases = all_kinases['GENE'].str.upper().drop_duplicates().to_list()
```

Split-explode approach

```python
prot = patients_proteins.copy()

# split-explode the Gene name column and filter for kinases
prot['Gene names'] = prot['Gene names'].str.split(';')
prot = prot.explode('Gene names')
prot = prot.loc[prot['Gene names'].isin(all_kinases)]
prot.groupby('Gene names').agg({patient: np.mean for patient in prot.filter(regex=patient_regex).columns.tolist()})
```

Unique peptide approach

```python
prot = patients_proteins.copy()

# filter out peptides potentially found in multiple Proteins and filter for kinases
prot = prot.loc[prot['Gene names'].str.count(';') == 0]
prot = prot.loc[prot['Gene names'].isin(all_kinases)]
prot.groupby('Gene names').agg({patient: np.mean for patient in prot.filter(regex=patient_regex).columns.tolist()})
```

Occam's razor approach

```python
prot = patients_proteins.copy()

# count the number of unique peptides per kinase
kinase_values = prot.loc[prot['Gene names'].isin(all_kinases)]['Gene names']
kinase_values = kinase_values.value_counts()
```

```python
def find_most_occurring_kinase(kinase_string):
    if type(kinase_string) != str:
        raise ValueError(f'Unexpected value {kinase_string} of type {type(kinase_string)}')
    kinase_string = kinase_string.split(';')
    kinase_string = [kinase for kinase in kinase_string if kinase in kinase_values.index]
    if len(kinase_string) == 0:
        return np.NaN
    else:
        max = 0
        for kinase in kinase_string:
            if kinase_values[kinase] > max:
                max = kinase_values[kinase]
                max_kinase = kinase
        return max_kinase
```

```python
amb = prot['Gene names']
amb = amb.loc[amb.str.count(';') != 0]
amb = amb.loc[~amb.isna()]
amb
```

```python
pre_dict = pd.concat([amb, amb.apply(find_most_occurring_kinase).rename('into')], axis=1).drop_duplicates()
repl_dict = pd.Series(pre_dict['into'].values, index=pre_dict['Gene names']).to_dict()
```

```python
prot = patients_proteins.copy()
prot = prot.loc[prot['Gene names'].isin(all_kinases)]
prot['Gene names'] = prot['Gene names'].replace(repl_dict)
prot.groupby('Gene names').agg({patient: np.mean for patient in prot.filter(regex=patient_regex).columns.tolist()})
```
